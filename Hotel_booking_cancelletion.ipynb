Country


There are 488 missing values for country, accounting for 0.4% of the data. We can't impute the values for these rows from the other columns. The remaining options would be to set these to Unknown (though this could introduce bias) or drop the rows entirely.

Another thing to consider is that there are 177 unique values for country; this will be too many to convert to dummy variables later. We can groups these by continent to reduce dimensionality in our data.

First, lets drop the rows with missing data, then define and execute a function to create a new continent column based on the country codes.

```python
df = df.dropna(subset=['country'])
```

```python
df['country'].value_counts()
```

```python
import pycountry
import pycountry_convert as pc
```

```python
def convert_country_code(country):
    alpha2_code = None
    try:
        if len(country) == 3:
            country_obj = pycountry.countries.get(alpha_3=country)
            if country_obj is None:
                return "Invalid code"
            alpha2_code = country_obj.alpha_2
        elif len(country) == 2:
            country_obj = pycountry.countries.get(alpha_2=country)
            if country_obj is None:
                return "Invalid code"
            alpha2_code = country
        continent = pc.country_alpha2_to_continent_code(alpha2_code)
        return continent
    except KeyError:
        return "Invalid code"
```

```python
df['continent'] = df['country'].apply(convert_country_code)
```

```python
df['continent'].value_counts()
```

```python
df[df['continent']=='Invalid code']['country']
```

```python
df.loc[df['continent']=='Invalid code', 'continent'] = df['country'].map({'TMP':'AS','UMI':'OC','ATA':'OC','ATF':'OC'})
```

The above columns can be converted into dummy variables except for four columns:

assigned_room_type as it is multicollinear with reserved_room_type, and also because we will be predicting only using information available prior to arrival

reservation_status as this is similar to the dependent variable is_canceled with the addtion of No-Show. which will be unknown prior to arrival

reservation_status_date as it is not relevant

country as we will be predicting off the continent column

```python
df = df.drop(['assigned_room_type','reservation_status','reservation_status_date','country'],axis=1)
```

```python
dummies = pd.get_dummies(df[['hotel','arrival_date_month','meal','market_segment','distribution_channel','reserved_room_type','deposit_type','customer_type','continent']],drop_first=True)
```

```python
df = pd.concat([df.drop(['hotel','arrival_date_month','meal','market_segment','distribution_channel','reserved_room_type','deposit_type','customer_type','continent'],axis=1),dummies],axis=1)
```

We can also drop the contact details and credit card number as they consist of unique values and are not useful predictors.

```python
df = df.drop(['name','email','phone-number','credit_card'],axis=1)
```

# Unnecessary Data

```python
df.head()
```

We have already dropped all the unnecessary non-numeric data; here we will also be removing the arrival_date_year,arrival_date_week_number and arrival_date_day_of_month.

The year is not useful since we will be making predictions for bookings in the future. Additionally, we cannot make predictions using the continuous data for week number and day of month, as a linear relationship would not exist between these variables and cancellations. The categorical dummies created for arrival_date_month will be useful for making predictions based of the time of year.

```python
df = df.drop(['arrival_date_year','arrival_date_week_number','arrival_date_day_of_month'],axis=1)
```

# Splitting the Data

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('is_canceled',axis=1).values
```

```python
y = df['is_canceled'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

# Random Forest Model
RFC is well-suited for datasets like this that involve a mix of categorical and numerical features.

Creating the Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=300)
```

```python
rfc.fit(X_train,y_train)
```

Evaluating the Model

```python
predictions = rfc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_test,predictions)
```

```python
print(classification_report(y_test,predictions))
```

# Deep Learning Model
Below is a TensorFlow model which uses a neural network to predict hotel cancellations. It has an F1 score of 0.82, so it doesn't perform as well as the RFC (0.86).

Normalising the Data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

# Creating the Model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
model = Sequential()
```

```python
X_train.shape
```

```python
model.add(Dense(units=60,activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(units=30,activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(units=15,activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(units=1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=3)
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),batch_size=64,epochs=50,callbacks=[early_stopping])
```

## Evaluating the Model

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
tf_predictions = model.predict(X_test)
```

```python
tf_predictions = np.round(tf_predictions).astype(int)
```

```python
confusion_matrix(y_test,tf_predictions)
```

```python
print(classification_report(y_test,tf_predictions))
```
